# Example Breakdown

Take this line:

`0:[ASTV<59] yes=1,no=2,missing=2`


It means:

* Feature: ASTV

* Split condition: < 59

* If true → go to node 1

* If false → go to node 2

* If missing → go to node 2

Then later you see:

`35:leaf=-0.12394247`


That’s a leaf value, which is part of the trained parameters (learned weight for that path).
